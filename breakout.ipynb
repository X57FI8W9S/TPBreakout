```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar monitor.csv 
df = pd.read_csv("monitor.csv", comment="#")  # columnas: r (recompensa), l (longitud), t (tiempo)
df["ep"] = range(len(df))
ma = df["r"].rolling(50, min_periods=1).mean()

# 1) Recompensa vs Episodio (+ media móvil)
plt.figure(figsize=(8,4.5))
plt.plot(df["ep"], df["r"], alpha=0.25, label="Recompensa (episódica)")
plt.plot(df["ep"], ma, label="Media móvil (50 ep)")
plt.xlabel("Episodio"); plt.ylabel("Recompensa"); plt.title("Breakout DQN — Recompensa")
plt.legend(); plt.tight_layout(); plt.show()

# 2) Duración del episodio vs Episodio
plt.figure(figsize=(8,4.5))
plt.plot(df["ep"], df["l"], alpha=0.8)
plt.xlabel("Episodio"); plt.ylabel("Pasos"); plt.title("Duración del episodio")
plt.tight_layout(); plt.show()

# 3) Distribución de recompensas
plt.figure(figsize=(7,4.2))
df["r"].hist(bins=30)
plt.xlabel("Recompensa por episodio"); plt.ylabel("Cantidad"); plt.title("Distribución de recompensas")
plt.tight_layout(); plt.show()
```
